```python
# Scale the variables

X_scale = preprocessing.scale(X)
X_scale[:10]
```

```python
# Finding the optimum number of clusters for k-means classification

x = df.iloc[:, [0, 1, 2, 3]].values

from sklearn.cluster import KMeans
css = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    css.append(kmeans.inertia_)
```

# Elbow Curve

```python
plt.plot(range(1,11), css)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('CSS') # cluster sum of squares ( inside the cluster )
plt.show()
```

# Based on Curve we can plot our graph with 2 , 3 and 5 no. of clusters

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
# 2 cluster

kmeans_2 = KMeans(2)
kmeans_2.fit(X_scale)

cul_2 = X.copy()

cul_2['pred'] = kmeans_2.fit_predict(X_scale)

plt.figure(figsize=(10,6))
plt.scatter(cul_2['SepalLengthCm'], cul_2['SepalWidthCm'], c= cul_2['pred'], cmap = 'rainbow');
```
